<h3><b>Task 1</b>

Необходимо написать программу с функцией multiplicate(A), принимающей на вход 
массив целых чисел А ненулевой длины и массив такой же длины, в котором на i-ом 
месте находится произведение всех чисел массива А, кроме числа, стоящего на i-ом 
месте.

```python
def multiplicate(A):
    if len(A) == 0:
        return None
    # multiplication of all elements in list
    a_mult = 1
    # number of zeros in array
    n_zeros = 0
    for el in A:
        if el == 0:
            n_zeros += 1
            continue
        a_mult *= el
    # if number of zeros in list >= 2, we always will get 0 in result, so we don't need to multiply
    if n_zeros >= 2:
        return [0] * len(A)
    result = list()
    for el in A:
        if (el == 0) and (n_zeros == 1):
            result.append(a_mult)
        elif (el != 0) and (n_zeros == 1):
            result.append(0)
        else:
            result.append(a_mult / el)
    return result
```

<b> Complexity: 2 * len(A) 

```python
def test(A, exp_res):
    res = multiplicate(A)
    print(f"Expected result: {exp_res}")
    print(f"Result: {res}")
    print(exp_res == res)
```

```python
A =  [1, 2, 3, 4]
exp_res =  [24, 12, 8, 6]
test(A, exp_res)
```

```python
A =  [1, 2, 3, 4, 0]
exp_res =  [0, 0, 0, 0, 24]
test(A, exp_res)
```

```python
A =  [1, 0, 3, 4, 0]
exp_res =  [0, 0, 0, 0, 0]
test(A, exp_res)
```

```python
A =  [1, -2, 3, 4]
exp_res =  [-24, 12, -8, -6]
test(A, exp_res)
```

<h3><b> Task 2

Необходимо реализовать алгоритм, позволяющий определить, является ли аудиозапись зашумлённой или нет.

```python
import tensorflow as tf
from tensorflow.keras import *
import matplotlib.pyplot as plt
import PIL.Image as Image
import os
import numpy as np
from tqdm import tqdm
```

```python
train_path = "train.part1/train/train"
train_path_2 = "train.part2/train/train"
val_path = "val/val"
```

```python
def fix_size(data):
    res_arr = np.zeros(shape = [2048, 80])
    res_arr[0 : data.shape[0], 0 : data.shape[1]] = data
    return res_arr
```

```python
def load_data(path):
    clean_data = list()
    noisy_data = list()
    clean_path = os.path.join(path, 'clean')
    noisy_path = os.path.join(path, 'noisy')
    speakers = os.listdir(clean_path)
    for speaker in tqdm(speakers):
        fnames = os.listdir(os.path.join(clean_path, speaker))
        for fname in fnames:
            clean_fpath = os.path.join(clean_path, speaker, fname)
            noisy_fpath = os.path.join(noisy_path, speaker, fname)
            if os.path.exists(clean_fpath) and os.path.exists(noisy_fpath):
                clean_el = np.load(clean_fpath)
                noisy_el = np.load(noisy_fpath)
                clean_data.append(clean_el)
                noisy_data.append(noisy_el)
    return clean_data, noisy_data
```

```python
train_cl, train_ns = load_data(train_path)
```

```python
train_cl_lens = [len(el) for el in train_cl]
train_ns_lens = [len(el) for el in train_ns]
```

```python
max(train_cl_lens)
```

```python
train_x = list()
train_x.extend(train_cl)
train_x.extend(train_ns)
train_x = list(map(fix_size, train_x))
train_x = np.array(train_x)
```

```python
train_x.shape
```

```python
train_y = np.zeros(shape=[len(train_x), 2])
train_y[0 : len(train_cl), 0] = 1
train_y[len(train_cl) : , 1] = 1
```

```python
val_cl, val_ns = load_data(val_path)
```

```python
val_x = list()
val_x.extend(val_cl)
val_x.extend(val_ns)
val_x = list(map(fix_size, val_x))
val_x = np.array(val_x)
```

```python
val_y = np.zeros(shape=[len(val_x), 2])
val_y[0 : len(val_cl), 0] = 1
val_y[len(val_cl) : , 1] = 1
```

```python
inp = layers.Input(shape=[2048, 80])
layer = layers.Conv1D(filters = 64, kernel_size = 5, strides = 2)(inp)
layer = layers.ReLU()(layer)
layer = layers.BatchNormalization()(layer)
layer = layers.Conv1D(filters = 128, kernel_size = 5, strides = 2)(layer)
layer = layers.ReLU()(layer)
layer = layers.BatchNormalization()(layer)
layer = layers.Conv1D(filters = 128, kernel_size = 5, strides = 2)(layer)
layer = layers.ReLU()(layer)
layer = layers.BatchNormalization()(layer)
layer = layers.Conv1D(filters = 128, kernel_size = 5, strides = 2)(layer)
layer = layers.ReLU()(layer)
layer = layers.BatchNormalization()(layer)
layer = layers.GlobalAvgPool1D()(layer)
out_layer = layers.Dense(2, activation = 'softmax')(layer)
model = models.Model(inputs=inp, outputs = out_layer)
model.summary()
```

```python
model.compile(
    loss = losses.CategoricalCrossentropy(),
    optimizer = optimizers.Adam(learning_rate = 0.0001),
    metrics = ['accuracy']
)
```

```python
model.fit(train_x, train_y, validation_data=(val_x, val_y), batch_size = 8, verbose = 1, epochs = 30, shuffle=True)
```

```python
model.save('classification')
```

```python
preds = model.predict(val_x)
```

```python
y_pred = np.argmax(preds, axis = 1)
y_true = np.argmax(val_y, axis = 1)
```

```python
from sklearn import metrics
```

```python
precision = metrics.precision_score(y_true, y_pred)
recall = metrics.recall_score(y_true, y_pred)
f1 = metrics.f1_score(y_true, y_pred)
print(f"Val precision: {precision}, val recall: {recall}, val f1: {f1}")
```

```python
class estimator:
    _estimator_type = ''
    classes_= []
    
    def __init__(self, model, classes):
        self.model = model
        self._estimator_type = 'classifier'
        self.classes_ = classes
        
    def predict(self, X):
        y_prob= self.model.predict(X)
        y_pred = y_prob.argmax(axis=1)
        return y_pred

classifier = estimator(model, classes = ['clear', 'noise'])
```

<b>Необходимо реализовать алгоритм, позволяющий избавиться от посторонних шумов в звуковом сигнале. 


```python
train_cl = np.array(list(map(fix_size, train_cl)))
train_ns = np.array(list(map(fix_size, train_ns)))
val_ns = np.array(list(map(fix_size, val_ns)))
val_cl = np.array(list(map(fix_size, val_cl)))
```

```python
def create_conv_layer(inp_layer, filters = 64, kernel_size = 5, stride = 2):
    layer = layers.Conv1D(filters = filters, kernel_size = kernel_size, strides = stride, padding = 'same')(inp_layer)
    layer = layers.ReLU()(layer)
    layer = layers.BatchNormalization()(layer)
    return layer
```

```python
def create_transpose_layers(inp_layer, filters = 64, kernel_size = 3, stride = 2):
    layer = layers.Conv1DTranspose(filters = filters, kernel_size = kernel_size, strides = stride, padding = 'same')(inp_layer)
    layer = layers.ReLU()(layer)
    layer = layers.BatchNormalization()(layer)
#     layer = layers.Conv1D(filters = filters, kernel_size = 5, strides = 1)(layer)
#     layer = layers.ReLU()(layer)
#     layer = layers.BatchNormalization()(layer)
    return layer
```

```python
inp = layers.Input(shape=[2048, 80])
conv_1 = create_conv_layer(inp)
conv_2 = create_conv_layer(conv_1)
conv_3 = create_conv_layer(conv_2)
unconv_1 = create_transpose_layers(conv_3, kernel_size=3, stride = 2)
unconv_1 = layers.Concatenate()([unconv_1, conv_2])
unconv_2 = create_transpose_layers(unconv_1)
unconv_2 = layers.Concatenate()([conv_1, unconv_2])
unconv_3 = create_transpose_layers(unconv_2, filters= 80)
out_layer = layers.Add()([inp, unconv_3])
model = models.Model(inputs=inp, outputs = out_layer)
model.summary()
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.compile(
    loss = losses.MeanAbsoluteError(),
    optimizer = optimizers.Adam(learning_rate = 0.0001),
    metrics = ['MAE', 'MSE']
)
```

```python
model.fit(train_ns, train_cl, validation_data=(val_ns, val_cl), batch_size = 8, verbose = 1, epochs = 50)
```
